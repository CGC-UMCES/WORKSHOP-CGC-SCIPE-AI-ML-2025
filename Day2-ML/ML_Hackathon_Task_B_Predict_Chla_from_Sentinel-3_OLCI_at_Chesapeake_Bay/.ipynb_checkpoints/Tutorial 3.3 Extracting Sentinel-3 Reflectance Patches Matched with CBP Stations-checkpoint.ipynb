### 3.3.1 Introduction

In this step, we extract small reflectance image patches from Sentinel-3 OLCI Level-2 products that correspond to in-situ surface chlorophyll-a (Chla) measurements collected by the Chesapeake Bay Program (CBP).

Each extracted patch:
- Is centered on a CBP station location and aligned with the sample date  
- Includes 16 OLCI reflectance bands: Oa01–Oa12, Oa16–Oa18, and Oa21  
- Includes a binary cloud mask derived from the WQSF quality flag  
- Is saved as a compressed `.npz` file, containing the reflectance patch, cloud mask, Chla value, and metadata

📂 **Output folders:**
- Extracted patch files: `patches/` (saved as `.npz`)
- Patch metadata index: `patches/patch_index.csv`


### 3.3.2 Setup and Load CBP In-situ Data

We begin by setting paths, defining band files, and loading the preprocessed CBP in-situ dataset (`averaged_layer_S.csv`). We also prepare a patch index to track extracted samples and avoid duplicates.


```python
import os
import xml.etree.ElementTree as ET
import pandas as pd
import numpy as np
import xarray as xr
from datetime import timedelta

# === Settings ===
base_dir = "products"
cbp_csv = "CleanedData/averaged_layer_S.csv"
patch_radius = 3
patch_dir = "MatchedPatches"
os.makedirs(patch_dir, exist_ok=True)
index_file = os.path.join(patch_dir, "patch_index.csv")

band_files = [f"Oa{str(i).zfill(2)}_reflectance.nc" for i in range(1,13)] + \
             ["Oa16_reflectance.nc", "Oa17_reflectance.nc", "Oa18_reflectance.nc", "Oa21_reflectance.nc"]

extra_files = [
    "chl_nn.nc", "chl_oc4me.nc", "tsm_nn.nc", "trsp.nc", "par.nc", "iop_nn.nc", "w_aer.nc", "wqsf.nc"
]

# === Load CBP data ===
cbp_df = pd.read_csv(cbp_csv, parse_dates=["SampleDate"])
print(f"📊 CBP records: {len(cbp_df)} from {cbp_df['SampleDate'].min().date()} to {cbp_df['SampleDate'].max().date()}")

# === Load or create patch index ===
if os.path.exists(index_file):
    patch_idx = pd.read_csv(index_file)
    done_keys = set(zip(patch_idx.date, patch_idx.Latitude, patch_idx.Longitude))
    print(f"📝 Loaded patch index with {len(patch_idx)} entries")
else:
    patch_idx = pd.DataFrame(columns=["patch_id", "date", "Latitude", "Longitude", "Measure", "mask_frac", "file"])
    done_keys = set()
    print("📌 Starting fresh patch index")
```

### 3.3.3 Extract Reflectance Patches

For each CBP measurement:
- Find a matching OLCI product within ±2 days
- Locate the nearest pixel to the station coordinates
- Extract reflectance patches across 16 bands
- Apply a cloud mask and compute the cloud fraction
- Save patch and metadata in netcdf


```python
import os
import xml.etree.ElementTree as ET
import pandas as pd
import numpy as np
import xarray as xr
from datetime import timedelta

# === Settings ===
base_dir = "products"
cbp_csv = "CleanedData/averaged_layer_S.csv"
patch_radius = 3
patch_dir = "MatchedPatches"
os.makedirs(patch_dir, exist_ok=True)
index_file = os.path.join(patch_dir, "patch_index.csv")

band_files = [f"Oa{str(i).zfill(2)}_reflectance.nc" for i in range(1,13)] + \
             ["Oa16_reflectance.nc", "Oa17_reflectance.nc", "Oa18_reflectance.nc", "Oa21_reflectance.nc"]

extra_files = [
    "chl_nn.nc", "chl_oc4me.nc", "tsm_nn.nc", "trsp.nc", "par.nc", "iop_nn.nc", "w_aer.nc", "rtc.nc"
]

# === Load CBP data ===
cbp_df = pd.read_csv(cbp_csv, parse_dates=["SampleDate"])
print(f"📊 CBP records: {len(cbp_df)} from {cbp_df['SampleDate'].min().date()} to {cbp_df['SampleDate'].max().date()}")

# === Load or create patch index ===
if os.path.exists(index_file):
    patch_idx = pd.read_csv(index_file)
    done_keys = set(zip(patch_idx.date, patch_idx.Latitude, patch_idx.Longitude))
    print(f"📝 Loaded patch index with {len(patch_idx)} entries")
else:
    patch_idx = pd.DataFrame(columns=["patch_id", "date", "Latitude", "Longitude", "Measure", "mask_frac", "file"])
    done_keys = set()
    print("📌 Starting fresh patch index")

# === Helper functions ===
def find_nearest_pixel(lat, lon, lat_array, lon_array):
    dist = np.sqrt((lat_array - lat)**2 + (lon_array - lon)**2)
    return np.unravel_index(np.argmin(dist), dist.shape)

def extract_patch(arr, i, j, radius):
    return arr[i-radius:i+radius, j-radius:j+radius]

def create_cloud_mask(wqsf_arr):
    CLOUD_MASK = 1 << 1
    return (wqsf_arr & CLOUD_MASK) != 0

def find_matching_subset(product_dirs, target_date):
    for delta in range(0, 1):  ±0, #2: ±1, #3: ±2
        for offset in [-delta, delta]:
            try_date = target_date + timedelta(days=offset)
            for extract_dir in product_dirs:
                manifest_path = os.path.join(extract_dir, "xfdumanifest.xml")
                if not os.path.exists(manifest_path):
                    continue
                try:
                    tree = ET.parse(manifest_path)
                    ns = {'safe':'http://www.esa.int/safe/sentinel/1.1'}
                    start_time = tree.getroot().find('.//safe:acquisitionPeriod/safe:startTime', ns).text
                    rs_date = pd.to_datetime(start_time).normalize().date()
                    if rs_date == try_date:
                        return extract_dir, rs_date
                except:
                    continue
    return None, None

# === Begin patch extraction ===
product_dirs = [os.path.join(root) for root, dirs, files in os.walk(base_dir) if 'xfdumanifest.xml' in files]
print(f"📦 Found {len(product_dirs)} OLCI products")

added_total = 0
skipped_total = 0

for _, row in cbp_df.iterrows():
    date, lat, lon, meas = row.SampleDate.date(), row.Latitude, row.Longitude, row.MeasureValue
    if (str(date), lat, lon) in done_keys:
        continue

    extract_dir, used_date = find_matching_subset(product_dirs, date)
    if extract_dir is None:
        print(f"⏭️ No product found ±2 days for {date} at ({lat}, {lon})")
        continue

    try:
        geo = xr.open_dataset(os.path.join(extract_dir, "geo_coordinates.nc"))
        lat_arr, lon_arr = geo["latitude"].values, geo["longitude"].values

        wqsf = xr.open_dataset(os.path.join(extract_dir, "wqsf.nc"))["WQSF"].values
        cloud_mask = create_cloud_mask(wqsf)

        i, j = find_nearest_pixel(lat, lon, lat_arr, lon_arr)
        cloud_patch = extract_patch(cloud_mask, i, j, patch_radius)
        mask_frac = cloud_patch.mean() if cloud_patch.shape == (patch_radius*2, patch_radius*2) else np.nan

        stack = []
        for bf in band_files:
            fpath = os.path.join(extract_dir, bf)
            if not os.path.exists(fpath):
                raise FileNotFoundError(f"Missing band: {bf}")
            arr = xr.open_dataset(fpath).data_vars.values().__iter__().__next__().values
            patch = extract_patch(arr, i, j, patch_radius)
            if patch.shape != (patch_radius*2, patch_radius*2):
                raise ValueError(f"Bad shape in {bf}")
            stack.append(patch)

        for ef in extra_files:
            fpath = os.path.join(extract_dir, ef)
            if os.path.exists(fpath):
                arr = xr.open_dataset(fpath).data_vars.values().__iter__().__next__().values
                patch = extract_patch(arr, i, j, patch_radius)
                if patch.shape == (patch_radius*2, patch_radius*2):
                    stack.append(patch)

        cube = np.stack(stack, axis=-1)

        # === Save as NPZ ===
        patch_id = f"{used_date}_{lat:.6f}_{lon:.6f}"
        file_name = patch_id + ".npz"
        output_path = os.path.join(patch_dir, file_name)

        np.savez_compressed(output_path, patch=cube, cloud_mask=cloud_patch, mask_frac=mask_frac, measure=meas)

        patch_idx = pd.concat([patch_idx, pd.DataFrame([{
            "patch_id": patch_id,
            "date": used_date,
            "Latitude": lat,
            "Longitude": lon,
            "Measure": meas,
            "mask_frac": mask_frac,
            "file": file_name
        }])], ignore_index=True)

        added_total += 1
        if added_total % 10 == 0:
            print(f"✅ Saved {added_total} patches")

    except Exception as e:
        print(f"⚠️ Failed for {date} ({lat:.4f},{lon:.4f}): {e}")
        skipped_total += 1

# Save patch index
patch_idx.to_csv(index_file, index=False)
print(f"\n📅 Index saved: {index_file}")
print(f"✅ Total saved: {added_total} | ❌ Skipped: {skipped_total}")
```

### 3.3.4 Visual Verification of Extracted Patches

Let’s load one exported .npz patch and visualize selected bands (e.g., Oa03, Oa08, Oa12) along with the cloud mask to verify quality and alignment.


```python
import matplotlib.pyplot as plt
import numpy as np
import random
import os

# === Pick a random patch from output directory
patch_dir = "MatchedPatches"
patch_files = [f for f in os.listdir(patch_dir) if f.endswith(".npz")]
sample_patch = random.choice(patch_files)
print(f"🎯 Sample patch: {sample_patch}")

# === Load the NPZ file
data = np.load(os.path.join(patch_dir, sample_patch))
cube = data["patch"]  # shape: (6, 6, 16)
cloud_mask = data["cloud_mask"]

# === Select bands (Oa03 ~ index 2, Oa08 ~ index 7, Oa12 ~ index 11)
band_indices = {"Oa03": 2, "Oa08": 7, "Oa12": 11}

# === Plot selected bands and cloud mask
fig, axes = plt.subplots(1, 4, figsize=(16, 4))
for ax, (band_name, idx) in zip(axes[:3], band_indices.items()):
    im = ax.imshow(cube[:, :, idx], cmap='viridis')
    ax.set_title(f"{band_name} Reflectance")
    plt.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
    ax.axis('off')
import matplotlib.colors as mcolors

# Create a binary colormap
binary_cmap = mcolors.ListedColormap(['black', 'white'])  # 0 = black (cloud), 1 = white (clear)
bounds = [0, 0.5, 1]
norm = mcolors.BoundaryNorm(bounds, binary_cmap.N)

# === Cloud mask overlay
ax = axes[3]
im = ax.imshow(cloud_mask, cmap=binary_cmap, norm=norm)
ax.set_title("Cloud Mask (0 = cloud, 1 = clear)")

# Binary colorbar
cbar = plt.colorbar(im, ax=ax, ticks=[0, 1], fraction=0.046, pad=0.04)
cbar.ax.set_yticklabels(["Cloud", "Clear"])
ax.axis('off')

plt.suptitle(f"Visual Check — {sample_patch}", fontsize=14)
plt.tight_layout()
plt.show()
```
