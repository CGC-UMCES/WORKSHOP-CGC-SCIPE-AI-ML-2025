# üèÅ Task A ‚Äî Forecast Dissolved Oxygen (DO) at Station CB8.1

In this task, you will build models to **forecast Dissolved Oxygen (DO)** at the **next sampling date** for Station **CB8.1**.  

- Each row in the dataset corresponds to one sampling date (roughly every 14 days).  
- use measurements from the current sampling (and optionally past history) to predict DO at the following sampling event**. 

üìÇ **Dataset:** `CleanedData/TrackA_Data_CB8dot1.csv`

---

## üìä Dataset Summary
- **Number of rows (records):** 512  
- **Number of columns (target and predictors(features)):** 28  
- **Time coverage:** 1985-11-12 ‚Üí 2024-12-10  
- **Typical interval:** ~14 days  

## üîë Columns we have
- `Station` ‚Üí Station code (always CB8.1 here)  
- `Date` ‚Üí Sampling date  
- `DO` ‚Üí Dissolved oxygen (mg/L) **üéØ Target variable** 
    
### Features(Predictors) from Chesapeake Bay Program (CBP) ‚Äî Water Quality (Surface Data)
- `CHLA` ‚Üí Chlorophyll-a (¬µg/L)  
- `WTEMP` ‚Üí Water temperature (¬∞C)  
- `SALINITY` ‚Üí Salinity (PSU)  
- `TN` ‚Üí Total nitrogen (mg/L)  
- `TP` ‚Üí Total phosphorus (mg/L)  

### Features(Predictors) from USGS ‚Äî River Discharge (cubic feet per second)
- **Daily discharge flows:**  
  - `James_discharge`, `Potomac_discharge`, `Susquehanna_discharge`  
- **Accumulated discharges (historical flow memory):**  
  - `*_acc_flow_30d` ‚Üí 30-day cumulative discharge  
  - `*_acc_flow_60d` ‚Üí 60-day cumulative discharge  
  - `*_acc_flow_90d` ‚Üí 90-day cumulative discharge
